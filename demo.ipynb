```python
import pandas as pd

from LPatternIdentification import feature_set, split_data, get_patterns_from_feature_set, reduce_pattern_set
```

```python
# Load dataset
df = pd.read_csv('USPresidency.csv', index_col=0)
df.head(10)
```

```python
# Separate observations and labels
observations = df.iloc[:,:-1].to_numpy(dtype='int8')
labels = list(df['Target'])
```

```python
# Find set of relevant features
elections_feature_set = feature_set(observations, labels)
elections_feature_set
```

```python
# Split observations and labels by class
split_point, Good, Bad = split_data(observations, labels)
```

```python
# Return set of feature patterns corresponding to Good
Patterns = get_patterns_from_feature_set(Good, elections_feature_set)
Patterns
```

```python
# Reduce patterns to uniquely Good patterns
Patterns_reduced = reduce_pattern_set(Patterns, Bad, 7)
Patterns_reduced 
```
